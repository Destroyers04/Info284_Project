```python
df = pd.DataFrame(dataset)

print(df.columns)
print(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({'font.size': 8})

#main area start & stop
sns.scatterplot(y = 'Hovedområde start', x = 'Hovedområde stopp', data=df)
plt.title('main area start/stop')
plt.show()
sns.scatterplot(y = 'Art FAO', x = 'Hovedområde start',data=df)
plt.title('fish species & main area start')
plt.rcParams['figure.figsize'] = [18, 18]
plt.show()
```

- main area start is unique to main area stop.
- how to show the density of each point in a scatter graph? Want to see how many times each speices has been caught in each main area 

# sea depth

```python
#sea depth start and stop
sns.scatterplot(x = 'Havdybde start', y = 'Havdybde stopp', data=df)
plt.title('sea depth')
plt.show()
```

```python
#sea depth start and fish speices
sns.scatterplot(x = 'Havdybde start', y = 'Art FAO', data=df)
plt.title('sea depth start')
plt.rcParams['figure.figsize'] = [18, 18]
plt.show()
#sea depth stop and fish speices
sns.scatterplot(x = 'Havdybde stopp', y = 'Art FAO', data=df)
plt.title('sea depth stop')
plt.rcParams['figure.figsize'] = [18, 18]
plt.show()
```

# fishing equipment('Redskap FAO')

```python
sns.scatterplot(x = 'Redskap FAO', y = 'Art FAO', data=df)
plt.title('gear and species')
plt.rcParams['figure.figsize'] = [15, 18]
plt.show()
```

# time start / stop

```python
#time start and stop
sns.scatterplot(x = 'Starttidspunkt', y = 'Stopptidspunkt', data=df)
plt.title('time start and stop')
plt.show()
```

# boat length group ('Lengdegruppe')

```python
plt.rcParams['figure.figsize'] = [5, 18]
sns.scatterplot(x = 'Lengdegruppe', y = 'Art FAO (kode)', data=df)
plt.title('time start and stop')
plt.show()
```

```python
#heatmap

plt.rcParams['figure.figsize'] = [50, 50]
heatmap_data = df.groupby(['Art FAO (kode)','Lengdegruppe']).size().unstack(fill_value=0)
sns.heatmap(heatmap_data, cmap='Reds', annot=True, fmt='d', square = True)
plt.title('Heatmap of Fish Species and boat length')

plt.show()
```
