I want to see the distribution of length in this feature, are there any length that appears more often.

```python
count_of_length_groups = dataset.groupby(["Lengdegruppe"])["Lengdegruppe"].count()
count_of_length_groups.sort_values(ascending=True, inplace=True)
print(count_of_length_groups)
```

Looks like the majority of catches are made by the bigger boats.
