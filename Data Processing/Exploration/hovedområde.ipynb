to check if location is important we will have to see the distribution of areas when a species is caught as bycatch and when it isnt caught as a bycatch

```python
import warnings # Got an irritating warning
warnings.filterwarnings('ignore')
import pandas as pd
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("H:\Informasjonsvitenskap\Programming\Python\Info-284\Info284_Project\Exam Task\Dataset\elektronisk-rapportering-ers-2018-fangstmelding-dca-simple.csv", sep=";")

# Dataset where the species isn't the same as the main-species
bycatch = dataset[dataset['Art FAO'] != dataset['Hovedart FAO']]
torsk_only = dataset[dataset['Art FAO'] == "Torsk"]
torsk_only_Bycatch = torsk_only[torsk_only['Hovedart FAO'] != "Torsk"]
torsk_only_main = torsk_only[torsk_only['Hovedart FAO'] == "Torsk"]
```

```python

# Count the occurrences of each unique value in 'Hovedomrde start', sort, and select top 10
top_10_areas_bycatch_start = torsk_only_Bycatch['Hovedomrde start'].value_counts().head(10)

# Plot the top 10 areas
top_10_areas_bycatch_start.plot(kind='pie', figsize=(12, 10), fontsize=10, autopct='%1.1f%%', startangle=140)

# Setting the title and labels for clarity with smaller font size
plt.title('Top 10 Torsk Bycatch Main Start Areas', fontsize=12)
plt.ylabel('')
# Display the plot
plt.legend(title='Main Start Areas', loc='upper right', bbox_to_anchor=(0, 1))
plt.show()
```

```python
# Count the occurrences of each unique value in 'Hovedomrde start', sort, and select top 10
top_10_areas_bycatch_stop = torsk_only_Bycatch['Hovedomrde stopp'].value_counts().head(10)

# Plot the top 10 areas
top_10_areas_bycatch_stop.plot(kind='pie', figsize=(12, 10), fontsize=10, autopct='%1.1f%%', startangle=140)

# Setting the title and labels for clarity with smaller font size
plt.title('Top 10 Torsk Bycatch Main Stop Areas', fontsize=12)
plt.ylabel('')
# Display the plot
plt.legend(title='Main Stop Areas', loc='upper right', bbox_to_anchor=(0, 1))
plt.show()
```

```python

# Count the occurrences of each unique value in 'Hovedomrde start', sort, and select top 10
top_10_areas = torsk_only_main['Hovedomrde start'].value_counts().head(10)

# Plot the top 10 areas
top_10_areas.plot(kind='pie', figsize=(12, 10), fontsize=10, autopct='%1.1f%%', startangle=140)

# Setting the title and labels for clarity with smaller font size
plt.title('Top 10 Torsk Non-Bycatch Main Start Areas', fontsize=12)
plt.ylabel('')
# Display the plot
plt.legend(title='Main Start Area', loc='upper right', bbox_to_anchor=(0, 1))
plt.show()
```

```python

# Count the occurrences of each unique value in 'Hovedomrde start', sort, and select top 10
top_10_areas = torsk_only_main['Hovedomrde stopp'].value_counts().head(10)

# Plot the top 10 areas
top_10_areas.plot(kind='pie', figsize=(12, 10), fontsize=10, autopct='%1.1f%%', startangle=140)

# Setting the title and labels for clarity with smaller font size
plt.title('Top 10 Torsk Non-Bycatch Main Stop Areas', fontsize=12)
plt.ylabel('')
# Display the plot
plt.legend(title='Main Stop Areas', loc='upper right', bbox_to_anchor=(0, 1))
plt.show()
```
