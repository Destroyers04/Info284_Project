Unique Bycatch Counts per Main Species 

```python
unique_bycatch_counts = dataset.groupby('Hovedart FAO')['Art FAO'].nunique()

plt.figure(figsize=(20, 10))  
plt.rcParams.update({'font.size': 10})

unique_bycatch_counts.sort_values(inplace=True) 
unique_bycatch_counts.plot(kind='bar')

plt.title('Unique Bycatch Counts per Main Species')  
plt.xlabel('Main Species (Hovedart FAO)') 
plt.ylabel('Unique Bycatch Count')  
plt.xticks(rotation=45, ha="right")  
plt.xticks
plt.show()
```

```python
total_roundweight_by_main_species = dataset.groupby('Hovedart FAO')['Rundvekt'].sum()

total_roundweight_by_main_species.sort_values(inplace=True)


plt.figure(figsize=(20, 10))  
total_roundweight_by_main_species.plot(kind='bar')

plt.title('Total Round Weight of Bycatch per Main Species') 
plt.xlabel('Main Species (Hovedart FAO)') 
plt.ylabel('Total Round Weight')
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
