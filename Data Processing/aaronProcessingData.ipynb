```python
# Calculate the average 'Rundvekt' for each 'Art - gruppe'
df_average = df.groupby('Art - gruppe')['Rundvekt'].mean().reset_index()

# Sorting the results for better visualization
df_average_sorted = df_average.sort_values('Rundvekt', ascending=False)

# Plotting the average 'Rundvekt' for each 'Art - gruppe'
plt.figure(figsize=(10, 8))
ax = df_average_sorted.plot(kind='bar', x='Art - gruppe', y='Rundvekt', legend=None)

plt.xlabel('Art - gruppe')
plt.ylabel('Average Rundvekt')
plt.title('Average Round Weight Catch by Species Group')

# Format y-axis labels to avoid scientific notation and improve readability
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

#Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()
```
